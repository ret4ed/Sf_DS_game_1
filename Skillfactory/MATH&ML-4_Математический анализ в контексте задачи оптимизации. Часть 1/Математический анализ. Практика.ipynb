## Область определения

```python
from sympy import Symbol, S
from sympy.calculus.util import continuous_domain 
x = Symbol("x") 
f = (x**3)/(x**2-1) 
continuous_domain(f, x, S.Reals)
```

## Область значений функции

```python
from sympy import Symbol, S
from sympy.calculus.util import function_range
x = Symbol("x") 
f = (x**3)/(x**2-1) 
function_range(f, x, S.Reals)
```

## Нахождение точек пересечения с осями координат

```python
from sympy import solveset, Eq
f = (x**3)/(x**2-1) 
solveset(Eq(f, 0), x)
```

```python
x = Symbol("x")
f = (x**3)/(x**2-1)
f.subs(x, 0)
```

## Нахождение производной

```python
import sympy
from sympy import diff
x = sympy.Symbol("x")
expr = (x**3)/(x**2-1)
expr.diff(x)
```

```python
y = expr.diff(x)
solveset(Eq(y, 0), x)
```

## Нахождение второй производной

```python
import sympy
from sympy import diff
x = sympy.Symbol("x")
expr = (x**3)/(x**2-1)
expr.diff(x,2)
```

```python
y = expr.diff(x,2)
solveset(Eq(y, 0), x)
```

## Нахождение асимптот

```python
from sympy import Symbol, limit, oo
x = Symbol("x")
f = (x**3)/(x**2-1)
limit(f,x,oo)
```

```python
from sympy import Symbol, limit, oo
x = Symbol("x")
f = (x**3)/(x**2-1)
limit(f,x,1)
```

```python
from sympy import Symbol, limit, oo
x = Symbol("x")
f = (x**3)/(x**2-1)
limit(f,x,-1)
```

```python
from sympy import Symbol, limit, oo
x = Symbol("x")
f = (x**3)/((x**2-1)*x)
limit(f,x,oo)
```

```python
from sympy import Symbol, limit, oo
x = Symbol("x")
f = (x**3)/((x**2-1)) - 1*x
limit(f,x,oo)
```
