## Множества. Операции над множествами

### Операции над множествами Python



```python
A = {'ID453', 'ID312', 'ID41', 'ID3', 'ID500', 'ID920', 'ID36', 'ID27'}
B = {'ID41', 'ID36', 'ID27', 'ID124', 'ID7', 'ID501', 'ID91' }
```

#### Объединение множеств

```python
#Объединение множеств
union_AB = A.union(B)
print(union_AB)
```

```python
#Метод union можно заменить |
union_AB = A|B
print(union_AB)
```

#### Пересечение можеств

```python
#Пересечение множеств
inter_AB = A.intersection(B)
print(inter_AB)
```

```python
#Метод intersection можно заменить &
inter_AB = A&B
print(inter_AB)
```

#### Разность множеств


```python
#Разность множеств
diff_AB = A.difference(B)
print(diff_AB)
```

```python
#Метод difference можно заменить -
diff_AB = A-B
print(diff_AB)
```

```python
diff_BA = B-A
print(diff_BA)
```

#### Симметричная разность

```python
#Симметричная разность
symmAB = A.symmetric_difference(B)
print(symmAB)
```

```python
#Метод symmetric_difference можно заменить ^
symmAB = A^B
print(symmAB)
```

### Библиотека sympy, символьные вычисления и множества



```python
from sympy import FiniteSet, Union
l1 = [1, 3, 7, 10]
l2 = [9, 7, 1]
a = FiniteSet(*l1)
b = FiniteSet(*l2)
Union(a, b)
```

```python
from sympy import ConditionSet, Eq, Symbol, Interval
x = Symbol('x')
s = ConditionSet(x, Eq(x**2-5*x, 0), Interval(2,9))
s
```

## Исследование функции

### Область определения функции


```python
#Импортируем нужные функции для обозначения переменных
from sympy import Symbol, S 
#Импортируем функцию для поиска области значений
from sympy.calculus.util import function_range
#Импортируем функцию для поиска области определения
from sympy.calculus.util import continuous_domain
#Определяем нашу пременную 
x = Symbol('x')
#Определяем нашу функцию
f = 1/x
#Вычисляем область определения
continuous_domain(f, x, S.Reals)
```

```python
from sympy import log
x = Symbol('x')
f = log(x)/x
continuous_domain(f, x, S.Reals)
```

### Область значений функции


```python
#Импортируем нужные функции для обозначения переменных
from sympy import Symbol, S, sin
#Импортируем функцию для поиска области значений
from sympy.calculus.util import function_range
x =Symbol('x')
f = sin(x)
function_range(f, x, S.Reals)
```

```python
#Импортируем нужные функции для обозначения перменных
from sympy import Symbol, S
#Импортируем функцию для поиска области занчений
from sympy.calculus.util import function_range
#Добавляем функцию для вычисления экспоненциальной функции
from sympy import exp
x = Symbol('x')
f = (x*x-3)/(exp(x))
function_range(f, x, S.Reals)
```
