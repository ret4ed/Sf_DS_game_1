## Создание признаков

```python
#Загрузка датасета
import pandas as pd

df_wine = pd.read_csv('data/wine_cleared.csv')
display(df_wine.head())
```

### Разбор числовых велечин

```python
#Создадим новый признак price_round, 
#означающий округленную до целого числа цену за бутылку вина
#Для удобства сразу преобразуем признак в int
df_wine['price_round'] = df_wine['price'].round().astype(int)
display(df_wine.head())
```

### Разбор текста

```python
#Выполним код для нахождения года 
#вина при помощи регулярного выражения
regex = '\d{4}' #регулярное выражение для нахождения чисел
df_wine['year'] = df_wine['title'].str.findall(regex).str.get(0)
display(df_wine.head())
```

### Разбор категорий

```python
#Создадим новый бинарный признак is_usa,
#присвоим 1, если вино произведено в США, иначе - 0
df_wine['is_usa'] = df_wine['country'].apply(lambda x: 1 if x == 'US' else 0)
display(df_wine.head())
```

### Задание 2.1

```python
#Выбрать 2 самые популярные страны помимо США
display(df_wine['country'].value_counts())
```

### Задание 2.2

```python
#Создадим бинарные признаки is_france, is_italy
#В ответ вписать результат кода data['is_france'].sum()

df_wine['is_france'] = df_wine['country'].apply(lambda x: 1 if x == 'France' else 0)
df_wine['is_italy'] = df_wine['country'].apply(lambda x: 1 if x == 'Italy' else 0)

result_is_france = df_wine['is_france'].sum()
print(f'Результат выполнения кода для Франции: {result_is_france}')

result_is_italy = df_wine['is_italy'].sum()
print(f'Результат выполнения кода для Италии: {result_is_italy}')
```

### Задание 2.3

```python
#Создадим новый бинарный признак old_wine,
#где 1 означает, что вино старше 2010 года
#В ответ вписать результат кода data['old_wine'].sum()

df_wine['old_wine'] = df_wine['year'].apply(lambda x: 1 if float(x) < 2010 else 0)

result_old_wine = df_wine['old_wine'].sum()
print(f'Результат выполнения кода: {result_old_wine}')
```

```python
#Или так, это их вариант

# преобразуем признак year в объект datetime для удобного сравнения дат
df_wine['year'] = pd.to_datetime(df_wine['year'], errors='coerce')

# для сравнения используем год, заполняем значения признака old_wine, где год вина меньше 2010
df_wine['old_wine'] = df_wine['year'].apply(lambda x: 1 if x.year < 2010 else 0) 
result_old_wine = df_wine['old_wine'].sum()

print(f'Результат выполнения кода: {result_old_wine}')
```

### Задание 2.7

## Создание признаков. Работа с фактором "дата-время"

```python
import pandas as pd 

# инициализируем информацию о звонках
calls_list = [
    [460, '2013-12-17 04:55:39', '2013-12-17 04:55:44', '2013-12-17 04:55:45'],
    [12, '2013-12-16 20:03:20', '2013-12-16 20:03:22', '2013-12-16 20:07:13'],
    [56, '2013-12-16 20:03:20', '2013-12-16 20:03:20', '2013-12-16 20:05:04'],
    [980, '2013-12-16 20:03:20','2013-12-16 20:03:27', '2013-12-16 20:03:29'],
    [396, '2013-12-16 20:08:27', '2013-12-16 20:08:28','2013-12-16 20:12:03'],
    [449, '2013-12-16 20:03:20', '2013-12-16 20:03:25','2013-12-16 20:05:00'],
    [397, '2013-12-16 20:08:25', '2013-12-16 20:08:27', '2013-12-16 20:09:59'],
    [398, '2013-12-16 20:01:23', '2013-12-16 20:01:23', '2013-12-16 20:04:58'],
    [452, '2013-12-16 20:03:20', '2013-12-16 20:03:21','2013-12-16 20:04:55'],
    [440, '2013-12-16 20:03:20', '2013-12-16 20:04:26', '2013-12-16 20:04:32']
]

calls = pd.DataFrame(calls_list, columns= ['client_id', 'agent_date', 'created_at', 'end_date'])

#Преобразовываем признаки в формат datetime для удобной работы
calls['agent_date'] = pd.to_datetime(calls['agent_date'])
calls['created_at'] = pd.to_datetime(calls['created_at'])
calls['end_date'] = pd.to_datetime(calls['end_date'])

display(calls)
```

```python
#Посчитаем разницу между датой и временем начала 
#разговора с клиентом и датой и временем окончания звонка
calls['duration'] = (calls['end_date'] - calls['created_at']).dt.seconds
display(calls)
```

### Задание 4.1

```python
#Посчитать, сколько секунд тратят сотрудники компании на дозвон клиенту
calls['time_connection'] = (calls['created_at'] - calls['agent_date']).dt.seconds
display(calls)

sum_time_connection = calls['time_connection'].sum()
print(f'Общее время потраченное сотрудниками на дозвон клиенту: {sum_time_connection}')
```

### Задание 4.2

```python
#Создать новый признак определения состоявшегося звонка
calls['is_connection'] = calls['duration'].apply(lambda x: 1 if x >= 10 else 0)
display(calls)

sum_is_connection = calls['is_connection'].sum()
print(f'Общее количество состоявшихся звонков: {sum_is_connection}')
```

### Задание 4.3

```python
#Создать новый признак - разницу в секундах между 
#началом звонка оператора и его окончанием
calls['time_diff'] = (calls['end_date'] - calls['agent_date']).dt.seconds
display(calls)

sum_time_diff = calls['time_diff'].sum()
print(f'Общее время time_diff: {sum_time_diff}')
```

```python
#Удалим больше не нужные данные
calls = calls.drop(columns=['agent_date', 'created_at', 'end_date'], axis=1)
display(calls)
```

## Кодирование признаков. Методы

```python
#Установка и импрорт библиотеки для кодирования
import category_encoders as ce
```

```python
#Создадим набор для кодирования порядковых признаков
import pandas as pd
# инициализируем информацию об одежде
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list, columns=['size', 'type'])
display(clothing)
```

### Порядковое кодирование. Ordinal Encoding

```python
#Импортируем библиотеку для работы с кодировщиками 
import category_encoders as ce

#Создаем объект OrdinalEncoder, col - имя столбца, mapping - словарь с описанием кодировки
ord_encoder = ce.OrdinalEncoder(mapping=[{
    'col':'size',
    'mapping' : {'xxs': 1, 'xs': 2, 's': 3, 
                 'm': 4, 'l': 5, 'xxl': 6}
}])
#Применяем трансформацию к столбцу
data_bin = ord_encoder.fit_transform(clothing[['size']])
#Добавляем результат к исходному DataFrame
clothing = pd.concat([clothing, data_bin], axis=1) 

display(clothing)
```

### Задание 5.3

```python
#Вариант который предоставил ментор

df_wine_1 = pd.read_csv('data/wine_cleared.csv')
regex = '\d{4}' # регулярное выражение для нахождения чисел (ищем год)
df_wine_1['year'] = df_wine_1['title'].str.findall(regex).str.get(0)

ord_encoder = ce.OrdinalEncoder()
data_bin = ord_encoder.fit_transform(df_wine_1[['year']])
df_wine = pd.concat([df_wine, data_bin], axis=1)
df_wine
```

### Однократное кодирование. One-Hot Encoding

```python
#Закодируем признак type используя класс OneHotEncoding

#Импорт для работы с кодировщиком
import category_encoders as ce 

#Указываем столбец для кодирования
encoder = ce.OneHotEncoder(cols=['type'], use_cat_names=True)
type_bin = encoder.fit_transform(clothing['type'])
clothing = pd.concat([clothing, type_bin], axis=1)

display(clothing)
```

```python
#Реализация однократного кодирования в pandas
#при помощи функции pd.get_dummies()
clothing_dummies = pd.get_dummies(clothing, columns=['type'])
display(clothing_dummies)
```

### Задание 5.4

### Двоичное кодирование

```python
#Закодируем бинарным способом признак type 
#используя класс BinaryEncoder

#Импорт для работы с кодировщиком
import category_encoders as ce
#Указываем столбец для кодирования
bin_encoder = ce.BinaryEncoder(cols='type')
type_bin = bin_encoder.fit_transform(clothing['type'])
clothing = pd.concat([clothing, type_bin], axis=1)

display(clothing)
```
