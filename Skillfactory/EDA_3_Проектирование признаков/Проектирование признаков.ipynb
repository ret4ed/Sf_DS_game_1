## Создание признаков

```python
#Загрузка датасета
import pandas as pd

df_wine = pd.read_csv('data/wine_cleared.csv')
display(df_wine.head())
```

### Разбор числовых велечин

```python
#Создадим новый признак price_round, 
#означающий округленную до целого числа цену за бутылку вина
#Для удобства сразу преобразуем признак в int
df_wine['price_round'] = df_wine['price'].round().astype(int)
display(df_wine.head())
```

### Разбор текста

```python
#Выполним код для нахождения года 
#вина при помощи регулярного выражения
regex = '\d{4}' #регулярное выражение для нахождения чисел
df_wine['year'] = df_wine['title'].str.findall(regex).str.get(0)
display(df_wine.head())
```

### Разбор категорий

```python
#Создадим новый бинарный признак is_usa,
#присвоим 1, если вино произведено в США, иначе - 0
df_wine['is_usa'] = df_wine['country'].apply(lambda x: 1 if x == 'US' else 0)
display(df_wine.head())
```

### Задание 2.1

```python
#Выбрать 2 самые популярные страны помимо США
display(df_wine['country'].value_counts())
```

### Задание 2.2

```python
#Создадим бинарные признаки is_france, is_italy
#В ответ вписать результат кода data['is_france'].sum()

df_wine['is_france'] = df_wine['country'].apply(lambda x: 1 if x == 'France' else 0)
df_wine['is_italy'] = df_wine['country'].apply(lambda x: 1 if x == 'Italy' else 0)

result_is_france = df_wine['is_france'].sum()
print(f'Результат выполнения кода для Франции: {result_is_france}')

result_is_italy = df_wine['is_italy'].sum()
print(f'Результат выполнения кода для Италии: {result_is_italy}')
```

### Задание 2.3

```python
#Создадим новый бинарный признак old_wine,
#где 1 означает, что вино старше 2010 года
#В ответ вписать результат кода data['old_wine'].sum()

df_wine['old_wine'] = df_wine['year'].apply(lambda x: 1 if float(x) < 2010 else 0)

result_old_wine = df_wine['old_wine'].sum()
print(f'Результат выполнения кода: {result_old_wine}')
```

```python
#Или так, это их вариант

# преобразуем признак year в объект datetime для удобного сравнения дат
df_wine['year'] = pd.to_datetime(df_wine['year'], errors='coerce')

# для сравнения используем год, заполняем значения признака old_wine, где год вина меньше 2010
df_wine['old_wine'] = df_wine['year'].apply(lambda x: 1 if x.year < 2010 else 0) 
result_old_wine = df_wine['old_wine'].sum()

print(f'Результат выполнения кода: {result_old_wine}')
```
