### Сбор нескольких элементов: собираем все ссылки на странице

```python
#Задаем адрес ресурса
url = 'https://en.wikipedia.org/wiki/List_of_programming_languages'
#Делаем GET-запрос к ресурсу
response = requests.get(url)
#Создаем объект BeautifulSoup
page = BeautifulSoup(response.text, 'html.parser')
#Ищем ссылку по тегу <a> и выводим ее на экран
print(page.find('a'))
```

```python
#Ищем все ссылки на странице и сохраняем в переменной links в виде списка
links = page.find_all('a')
#Выводим количество найденных ссылок
print(len(links))
```

```python
#Выводим ссылки с 500 по 509 включительно
print([link.text for link in links[500:510]])
```

```python
#Выводим ссылки с 1 по 9 включительно
print([link.text for link in links[0:10]])
```

```python
#Создаем список всех ссылок
list_links = [link.text for link in links]
print(list_links)
#Ищем индексы первого и последнего языка программирования
print(list_links.index('A.NET (A#/A sharp)'))
print(list_links.index('ZPL'))

#Создаем список ссылкок языков программирования
list_links_new = list_links[147:858]
print(list_links_new)

#Создаем список очищенных языков программирования
programming_languages = []
for i in list_links_new:
    if i != 'edit':
        programming_languages.append(i)
        
print()       
print('На сайте указана информация о', len(programming_languages), 'языках программирования')

```

### Сбор информации из групп

```python
#Получим id участников группы VK(официальной группы VK)

#Импортируем модуль requests
import requests
#Указываем свой сервисный токен
token = '78c99b0478c99b0478c99b04317be5e684778c978c99b041f183f87f1bc2d712676982b'
#Указываем адрес обращения
url = 'https://api.vk.com/method/groups.getMembers'
#Формируем строку параметров
params = {'group_id': 'vk', 'v': 5.95, 'access_token': token}
#Посылаем запрос
response = requests.get(url, params=params)
#Ответ сохраняем в переменной data в формате словаря
data = response.json()
#Выводим содержимое переменной data на экран
print(data)
```

```python
#Выводим на экран количество элементов словаря
print(len(data['response']['items']))
```

```python
#Загружаем в переменную информацию об id первых 20 пользоателей в виде списка
users_for_cheking = data['response']['items'][:20]
#Выводим перечень id первых 20 пользователей
print(users_for_cheking)
```

```python
#Используем count и offset, чтобы получить те же id по пять за раз

#Импортируем модуль requests
import requests
#Указываем свой сервисный токен
token = '78c99b0478c99b0478c99b04317be5e684778c978c99b041f183f87f1bc2d712676982b'
#Указываем адрес обращения
url = 'https://api.vk.com/method/groups.getMembers'

#Будем выгружать по 5 пользователей, начиная с того места,
#где закончили на предыдущей итерации
count = 5
offset = 0
user_ids = []
max_count = 20
while offset < max_count:
    print(f'Выгружаю {count} пользователей с offset = {offset}')
    params = {'group_id': 'vk', 'v': 5.95, 'count': count, 'offset': offset, 'access_token': token}
    response = requests.get(url, params=params)
    data = response.json()
    user_ids += data['response']['items']
    #Увеличиваем смещение на количество строк, которое мы уже выгрузили
    offset += count
print(user_ids)
```

```python
#Сравним списки полученные 2 способами
print(user_ids == users_for_cheking)
```

### Ограничение по частоте запросов

```python
#Воспользуемся библиотекой time и методом sleep, с помощью которого мы можем добавить паузу,
#например 0.5 секунд, после каждого запроса

#Импортируем модуль requests
import requests
#Импортируем модуль time
import time
#Указываем свой сервисный токен
token = '78c99b0478c99b0478c99b04317be5e684778c978c99b041f183f87f1bc2d712676982b'
#Указываем адрес обращения
url = 'https://api.vk.com/method/groups.getMembers'

count = 1000
offset = 0
user_ids = []
while offset < 5000:
    params = {'group_id': 'vk', 'v': 5.95, 'count': count, 'offset': offset, 'access_token': token}
    response = requests.get(url, params=params)
    data = response.json()
    user_ids += data['response']['items']
    offset += count
    print('Ожидаю 0.5 секунд...')
    time.sleep(0.5)
print(f'Цикл завершен, offset = {offset}')
```

### Постановка задачи

```python
#Напишем задачу
def task():
    print('Hello! I am a task!')
    return 
```

```python
#Вызываем метод, определяющий единицы измерения промежутков времени,
#через которые будет выполняться функция
schedule.every(15).minutes.do(task)
```

```python
#Запуск задачи каждый час
schedule.every(1).hour.do(task)
```

```python
#Каждые 5 секунд
schedule.every(5).seconds.do(task)
```

### Выполнение функции

```python
#Создание паузы
import time
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
#Итоговый код
import schedule

def task():
    print('Hello! I am a task!')
    return

schedule.every(5).seconds.do(task)

import time
while True:
    schedule.run_pending()
    time.sleep(1)
```
