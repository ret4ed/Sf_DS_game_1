## Практика: векторы

Задача (условие № 1)

Агентство «Рай в Шалаше» недавно сдало в аренду четыре квартиры в одном доме.

Общая площадь: 33, 42, 33 и 65 кв. м.
Жилая площадь: 24, 32, 20 и 48 кв. м.
Арендная плата: 39, 50, 38 и 77 тысяч рублей соответственно.
В том же доме появилась новая квартира общей площадью 50 кв. м, из которых жилая — 40 кв. м.

На основе имеющихся данных выберите оптимальную арендную плату для новой квартиры.

```python
#Ключом к решению является линейная зависимость векторов: 
#можно заметить, что вектор p является линейной комбинацией 
#векторов sобщ и sжил с коэффициентами w1 = 1 и w2 = 1/4

#Составим вектор коэффициентов 
w = (1, 1/4)
#Составим вектор новой квартиры
x = (50, 40)

p = np.dot(w,x)
display(p)
```

Задача (условие № 2)

Пусть во второй квартире обнаружились бытовые проблемы, и её пришлось сдать дешевле (за 48 460 рублей), а в первой и третьей квартирах был застеклённый балкон, и их удалось сдать дороже (за 40 000 и за 39 000 рублей соответственно).

На основе имеющихся данных выберите оптимальную арендную плату для новой квартиры.



Теперь вектор арендной платы не выражается линейно через векторы площадей.

Изобразим данные геометрически. Так как векторы четырёхмерные, мы не сможем получить корректное геометрическое решение. Однако можно получить условное графическое представление, которого нам будет достаточно для решения задачи.

Векторы общей и жилой площади лежат в некоторой плоскости, обозначенной синим цветом. Вектор p из первого условия задачи линейно выражается через них, а значит, лежит в той же плоскости. 

Вектор p_реал выразить через векторы площадей уже нельзя, а значит он не лежит в синей плоскости.




Но что это значит для нас?

Это значит, что среди всех возможных линейных комбинаций векторов общей и жилой площади наше изначальное  — наилучший прогноз арендной платы.

Потому что ортогональная проекция по определению является вектором с ближайшим расстоянием к вектору, который проецируется.

Другими словами, если другой информации нет, то 60 тысяч рублей, которые мы получили в первом условии задачи, — и есть наш лучший прогноз стоимости аренды новой квартиры.
