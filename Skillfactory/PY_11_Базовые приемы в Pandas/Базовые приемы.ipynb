```python
import pandas as pd
```

```python
#Вызываем обновленный дата фрейм
melb_data_ps = pd.read_csv('data/melb_data_ps.csv', sep=',')
display(melb_data_ps.head())
display(melb_data_ps.tail())
```

### Базовые операции со столбцами DataFrame

```python
#Копируем основной дата фрейм
melb_df = melb_data_ps.copy()
melb_df.head()
```

```python
#Удаляем из таблицы столбцы index и Coordinates
#melb_df = melb_df.drop((['index','Coordinates']), axis=1)
#melb_df.head()
```

```python
#Альтернативный метод удаления столбцов
melb_df.drop(['index','Coordinates'],axis=1,inplace=True)
melb_df.head()
```

```python
#Производим операцию сложения между столбцами таблицы
total_rooms = melb_df['Rooms'] + melb_df['Bedroom'] + melb_df['Bathroom']
display(total_rooms)
```

```python
#Найдем среднюю площадь одной комнаты
melb_df['MeanRoomsArea'] = melb_df['BuildingArea'] / total_rooms
display(melb_df['MeanRoomsArea'])
```

```python
#Находим коэффициент соотношения площади здания к площади участка
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRation'] = diff_area/sum_area
display(melb_df['AreaRation'])
```

### Работа с датами в DataFrame

```python
#Вывод столбца с датами продаж
display(melb_df['Date'])
```

```python
#Приведем дату в стандартный формат
melb_df['Date'] = pd.to_datetime(melb_df['Date'],
dayfirst=True)
display(melb_df['Date'])
```

```python
#Проведем анализ по годам в столбце Date
years_sold = melb_df['Date'].dt.year
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
```

```python
#Проведем анализ по месяцам
melb_df['MonthSale'] = melb_df['Date'].dt.month
melb_df['MonthSale'].value_counts(normalize=True)
```

```python
#Вычислим разницу во времени между продажей и выбранной нами датой
delta_days = melb_df['Date'] - pd.to_datetime('2016-01-01')
display(delta_days)
#Изменим формат данных из количества дней в числа дней
display(delta_days.dt.days)
```

```python
#Найдем возраст здания на момент пррдажи
melb_df['AgeBuilding'] = melb_df['Date'].dt.year - melb_df['YearBuilt']
display(melb_df['AgeBuilding'])
```

```python
melb_df = melb_df.drop('YearBuilt', axis=1)
melb_df.head()
```

### Создание и преобразование столбцов с помощью функций

```python
#Вычислим количество уникальных значений в столбце с адресами
print(melb_df['Address'].nunique())
```

```python
#Посмотрим структура адресов
print(melb_df['Address'].loc[177])
print(melb_df['Address'].loc[1812])
print(melb_df['Address'].loc[9001])
```

```python
#Создаем функцию чтобы выделить подтип улицы
def get_street_type(address):
    exclude_list = ['N', 'S', 'W', 'E']
    address_list = address.split(' ')
    street_type = address_list[-1]
    if street_type in exclude_list:
        street_type = address_list[-2]
    return street_type
```

```python
#Применим написанную функцию к 
street_types = melb_df['Address'].apply(get_street_type)
display(street_types)
```

```python
#Узнаем сколько теперь уникальных значений получилось
print(street_types.nunique())
```

```python
#Посмотрим частоту уникальных значений для каждого типа
display(street_types.value_counts())
```

```python
#Выведем 10 самых популярных подтипов улиц
popular_stypes = street_types.value_counts().nlargest(10).index
print(popular_stypes)
```

```python
street_types = street_types.apply(lambda x: 
x if x != 'Avenue' else 'Av')
street_types = street_types.apply(lambda x: 
x if x != 'Boulevard' else 'Bvd')
street_types = street_types.apply(lambda x: 
x if x != 'Parade' else 'Pde')
display(street_types.value_counts())
```

```python
#Создадим новый столбец с подтипами улиц
#Если подтип в топ 10 популярных, то выводим его название
#В противном случае выводится other
melb_df['StreetType'] = street_types.apply(lambda x: x if x
in popular_stypes else 'other')
display(melb_df['StreetType'])
```

```python
#Найдем сколько уникальных типов в столбце StreetType
print(melb_df['StreetType'].nunique())
```

```python
#Так как столбец Address больше не несет пользы, удаляем его
melb_df = melb_df.drop('Address', axis=1)
```

### Задание 4.2

```python
#Пишем функцию для определения является ли день выходным
def get_weekend(weekday):
    if weekday == 5:
        return 1
    elif weekday == 6:
        return 1
    else:
        return 0
```

```python
#Создаем новый столбик и применяем функцию ко всем строкам
melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)
```

```python
#Находим среднюю цену проданной недвижимости в выходные
round(melb_df[melb_df['Weekend'] == 1]['Price'].mean())
```
