```python
import pandas as pd
```

```python
#Вызываем обновленный дата фрейм
melb_data_ps = pd.read_csv('data/melb_data_ps.csv', sep=',')
display(melb_data_ps.head())
display(melb_data_ps.tail())
```

```python
#Копируем основной дата фрейм
melb_df = melb_data_ps.copy()
melb_df.head()
```

```python
#Удаляем из таблицы столбцы index и Coordinates
#melb_df = melb_df.drop((['index','Coordinates']), axis=1)
#melb_df.head()
```

```python
#Альтернативный метод удаления столбцов
melb_df.drop(['index','Coordinates'],axis=1,inplace=True)
melb_df.head()
```

```python
#Производим операцию сложения между столбцами таблицы
total_rooms = melb_df['Rooms'] + melb_df['Bedroom'] + melb_df['Bathroom']
display(total_rooms)
```

```python
#Найдем среднюю площадь одной комнаты
melb_df['MeanRoomsArea'] = melb_df['BuildingArea'] / total_rooms
display(melb_df['MeanRoomsArea'])
```

```python
#Находим коэффициент соотношения площади здания к площади участка
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRation'] = diff_area/sum_area
display(melb_df['AreaRation'])
```

### Работа с датами в DataFrame

```python
#Вывод столбца с датами продаж
display(melb_df['Date'])
```

```python
#Приведем дату в стандартный формат
melb_df['Date'] = pd.to_datetime(melb_df['Date'],
dayfirst=True)
display(melb_df['Date'])
```

```python
#Проведем анализ по годам в столбце Date
years_sold = melb_df['Date'].dt.year
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
```

```python
#Проведем анализ по месяцам
melb_df['MonthSale'] = melb_df['Date'].dt.month
melb_df['MonthSale'].value_counts(normalize=True)
```

```python
#Вычислим разницу во времени между продажей и выбранной нами датой
delta_days = melb_df['Date'] - pd.to_datetime('2016-01-01')
display(delta_days)
#Изменим формат данных из количества дней в числа дней
display(delta_days.dt.days)
```

```python
#Найдем возраст здания на момент пррдажи
melb_df['AgeBuilding'] = melb_df['Date'].dt.year - melb_df['YearBuilt']
display(melb_df['AgeBuilding'])
```

```python
melb_df = melb_df.drop('YearBuilt', axis=1)
melb_df.head()
```
