## Задание 1

Реализовать функцию (или тело функции), которая проверяет на изоморфность два слова.

Пояснение: строки s и t называются изоморфными, если все вхождения каждого символа строки s можно последовательно заменить другим символом и получить строку t. Порядок символов при этом должен сохраняться, а замена — быть уникальной. Так, два разных символа строки s нельзя заменить одним и тем же символом из строки t, а вот одинаковые символы в строке s должны заменяться одним и тем же символом.

### Пример:

s = 'paper' 

t = 'title' 

print(is_isomorphic(s, t))

### Вывод: 

True

Оценить оптимальность решения по времени и памяти и прикрепить текст кода.

```python
#Функция для проверки 2-х строк на изоморфность
def is_isomorphic(str1, str2): #-> bool:
    
    #Проверка на одинаковую длину строк
    if len(str1) != len(str2):
        return False
    
    #Словари для символов
    mapping_str1_to_str2 = {}
    mapping_str2_to_str1 = {}

    for i in range(len(str1)):
        char_str1 = str1[i]
        char_str2 = str2[i]
    
        #Проверка соответствия str1 -> str2
        if char_str1 in mapping_str1_to_str2:
            if mapping_str1_to_str2[char_str1] != char_str2:
                return False
        else:
            mapping_str1_to_str2[char_str1] = char_str2
            
        #Проверка соответствия str2 -> str1
        if char_str2 in mapping_str2_to_str1:
            if mapping_str2_to_str1[char_str2] != char_str1:
                return False
        else:
            mapping_str2_to_str1[char_str2] = char_str1
            
    return True
```

```python
s = 'paper' 
t = 'title' 
print(is_isomorphic(s, t))

a = 'aac'
b = 'bhr'
print(is_isomorphic(a, b))

a = 'aa'
b = 'bhr'
print(is_isomorphic(a, b))

a = 'aab'
b = 'bbc'
print(is_isomorphic(a, b))
```

```python
def is_isomorphic(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False

    # Словари для отображения символов: s → t и t → s
    mapping_s_to_t = {}
    mapping_t_to_s = {}

    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]

        # Проверяем соответствие s → t
        if char_s in mapping_s_to_t:
            if mapping_s_to_t[char_s] != char_t:
                return False
        else:
            mapping_s_to_t[char_s] = char_t

        # Проверяем соответствие t → s (обратное отображение)
        if char_t in mapping_t_to_s:
            if mapping_t_to_s[char_t] != char_s:
                return False
        else:
            mapping_t_to_s[char_t] = char_s

    return True
```

```python
s = 'paper' 
t = 'title' 
print(is_isomorphic(s, t))

a = 'aac'
b = 'bhr'
print(is_isomorphic(a, b))

a = 'aa'
b = 'bhr'
print(is_isomorphic(a, b))

a = 'aab'
b = 'bbc'
print(is_isomorphic(a, b))
```

## Задание 2
Реализовать функцию (или тело функции), которая находит единственное отсутствующее число из последовательности натуральных чисел 1,2,…,n.

### Пример:

nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]

print(missing_number(nums))

### Вывод: 

7

Оценить оптимальность решения по времени и памяти и прикрепить текст кода.

```python
#Функция нахождения единственного отсутсвующего числа в натуральной последовательности
def missing_number(nums):
    #Количество чисел в последовательности с учетом пропущенного
    n = len(nums) + 1
    #Ожидаемая сумма
    S = n * (n + 1) // 2
    #Фактическая сумма
    sum_nums = sum(nums)
    return S - sum_nums
```

```python
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]

print(missing_number(nums))
```

## Задание 3

Реализовать функцию (или тело функции), которая
при введении натурального числа n разбивает его на простые множители (представить его в виде произведения простых чисел).

### Пример:

n = 56

print(prime_factors(n))

### Вывод:

[2, 2, 2, 7]

Оценить оптимальность решения по времени и памяти и прикрепить текст кода.



```python
#Функция разбиения числа n на простые множители
def prime_factors(n):
    #Возврат пустого списка при условии, что
    if n <=1:
        return []
    
    list_factors = []
    divisor = 2
    
    #Проверяем делитель только до √n
    while divisor * divisor <= n:
        while n % divisor == 0:
            list_factors.append(divisor)
            n //= divisor
        #После делителя 2 проверяем только нечетные
        divisor += 1 if divisor == 2 else 2
    
    #Если остался простой множитель
    if n > 1:
        list_factors.append(n)
    
    return list_factors
```

```python
n = 56
print(prime_factors(n))
```
